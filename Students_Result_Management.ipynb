```python
import numpy as np
```

```python
# students =[]
students = []

while True:

    print("\nMENU")
    print("1.Store Student Result")
    print("2.Update Student Result")
    print("3.Retrieve Student Result")
    print("4.Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:

        print("Good, now you're ready to store Result")

        name = input("Enter the Name of Student:")
        roll_no = int(input("Enter Roll Number: "))

        print("**Marks**")

        english = int(input("Enter English Marks: "))
        maths = int(input("Enter Maths Marks: "))
        science = int(input("Enter Science Marks: "))
        history = int(input("Enter History Marks: "))
        geography = int(input("Enter Geography Marks: "))

        total = english + maths + science + history + geography

        percentage = total / 500 * 100

        if percentage >= 90:
            grade = "A"

        elif percentage >= 75:
            grade = "B"

        elif percentage >= 50:
            grade = "C"

        else:
            grade = "Fail"

        print("Student Scored:", percentage)
        print("Grade:", grade)

        student = {
            "name": name,
            "roll_no": roll_no,
            "english": english,
            "maths": maths,
            "science": science,
            "history": history,
            "geography": geography,
            "percentage": percentage,
            "grade": grade
        }

        students.append(student)

        print("**Result Saved Successfully**")

    elif choice == 2:

        print("Now you can update results")

        update_name = input("Enter student name to update: ")

        found = False

        for s in students:

            if s["name"] == update_name:

                found = True

                print("""
1.Name
2.English Marks
3.Maths Marks
4.Science Marks
5.History Marks
6.Geography Marks
""")

                update = int(input("Enter your choice: "))

                if update == 1:

                    new_name = input("Enter new name: ")
                    s["name"] = new_name

                    print("Name Updated Successfully")

                elif update == 2:

                    new_english = int(input("Enter new English marks: "))
                    s["english"] = new_english

                    print("English Marks Updated Successfully")

                elif update == 3:

                    new_maths = int(input("Enter new Maths marks: "))
                    s["maths"] = new_maths

                    print("Maths Marks Updated Successfully")

                elif update == 4:

                    new_science = int(input("Enter new Science marks: "))
                    s["science"] = new_science

                    print("Science Marks Updated Successfully")

                elif update == 5:

                    new_history = int(input("Enter new History marks: "))
                    s["history"] = new_history

                    print("History Marks Updated Successfully")

                elif update == 6:

                    new_geography = int(input("Enter new Geography marks: "))
                    s["geography"] = new_geography

                    print("Geography Marks Updated Successfully")

                total = (
                    s["english"] +
                    s["maths"] +
                    s["science"] +
                    s["history"] +
                    s["geography"]
                )

                s["percentage"] = total / 500 * 100

                if s["percentage"] >= 90:
                    s["grade"] = "A"

                elif s["percentage"] >= 75:
                    s["grade"] = "B"

                elif s["percentage"] >= 50:
                    s["grade"] = "C"

                else:
                    s["grade"] = "Fail"

                print("Updated Percentage:", s["percentage"])
                print("Updated Grade:", s["grade"])

        if found == False:
            print("Student Not Found")

    elif choice == 3:

        if len(students) == 0:
            print("No Records Found")

        else:

            for s in students:

                print("\n----- STUDENT RESULT -----")

                print("Name:", s["name"])
                print("Roll Number:", s["roll_no"])
                print("English:", s["english"])
                print("Maths:", s["maths"])
                print("Science:", s["science"])
                print("History:", s["history"])
                print("Geography:", s["geography"])
                print("Percentage:", s["percentage"])
                print("Grade:", s["grade"])

    elif choice == 4:

        print("Exit")
        break

    else:
        print("Invalid choice")
```
